```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
words = [".", "Man", "is", "great", "god"]
stoi = {w:i for i,w in enumerate(words)}
```

```python
stoi
```

```python
itos = {i: w for i, w in enumerate(words)}
itos
```

```python
x=[stoi['.'],stoi['Man'],stoi['is'],stoi['great'],stoi['god']]
y=[stoi['Man'],stoi['is'],stoi['great'],stoi['.'],stoi['is']]
```

```python
X=torch.tensor(x,dtype=torch.long)
Y=torch.tensor(y,dtype=torch.long)
X
```

```python
class llm(nn.Module):
    def __init__(self,vocab_size,emb_dim):
        super().__init()
        self.emb = nn.Embedding(vocab_size,emb_dim)
        self.linear = nn.linear()
```

```python
emb = nn.Embedding(5,3)
layer = nn.Linear(3,5)
parameters = [*emb.parameters(),*layer.parameters()] # *.parameters() is used for geeting elements into list otherwise they are generator objects
emb.weight
```

```python
for p in parameters:
    p.requires_grad=True
z = emb(X)
z
```

```python
logits = layer(z)
loss = F.cross_entropy(logits,Y)
loss
```

```python
for _ in range(500):
    z = emb(X)
    logits = layer(z)
    loss = F.cross_entropy(logits,Y)
    for p in parameters:
        p.grad=None
    loss.backward()
    lr = 0.1
    for p in parameters:
        p.data += -lr*p.grad
print(loss)
```

```python
probs = F.softmax(logits,dim=1)
```

```python
s =''
```

```python
st='god'
s = st
while st != '.':
    row_idx = words.index(st)
    pred_idx = torch.max(probs[row_idx], dim=0)[1].item()
    st = words[pred_idx]
    if st!='.':
        s += ' ' + st
    else :
        s +=st
```
