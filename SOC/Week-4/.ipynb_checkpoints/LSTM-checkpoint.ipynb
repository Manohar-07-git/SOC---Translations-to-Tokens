```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

While Vanilla Recurrent Neural Networks (RNNs) are theoretically capable of processing sequential data of any length, they suffer from severe limitations in practice. These drawbacks make them highly unstable to train and incapable of handling long-range dependencies.

---

### 1. The Vanishing Gradient Problem
This is the most notorious issue with vanilla RNNs. 
* **The Cause:** During Backpropagation Through Time (BPTT), gradients are calculated by repeatedly multiplying the recurrent weight matrix ($W_{hh}$) across every time step. If the eigenvalues of $W_{hh}$ are less than 1, the gradient shrinks exponentially as it travels backward.
* **The Consequence:** By the time the gradient reaches the earliest time steps in a long sequence, it becomes practically zero. The network completely "forgets" the early inputs and fails to learn long-term dependencies.

### 2. The Exploding Gradient Problem
The exact mathematical opposite of the vanishing gradient problem.
* **The Cause:** If the eigenvalues of the recurrent weight matrix $W_{hh}$ are greater than 1, repeated multiplication causes the gradients to grow exponentially as they propagate backward through time.
* **The Consequence:** This leads to massive weight updates, causing the training process to oscillate wildly or completely crash due to numerical overflow (resulting in `NaN` or `Inf` values).

### 3. Inability to Capture Long-Range Dependencies
Because of the vanishing gradient problem, standard RNNs effectively only have a short-term memory. 
* **Example:** In a language modeling task, if an RNN processes a long paragraph:
  > *"The **clouds** in the sky, which had been gathering all morning over the valley while we packed our bags, were **dark**."*
* The network will likely fail to connect the plural subject *"clouds"* to the later context because too many time steps (words) separate them.

### 4. High Computational Complexity & Lack of Parallelism
Unlike Feedforward Networks or Transformers, RNNs process data strictly sequentially.
* To calculate the hidden state $h_t$ at time step $t$, the network *must* wait for the completion of the hidden state $h_{t-1}$ from the previous time step.
* This sequential bottleneck prevents modern GPUs from parallelizing the workload across time steps, making training on long sequences painfully slow.

---

### Summary of Solutions
To overcome these structural flaws, modern deep learning relies on upgraded architectures:
* **LSTMs & GRUs:** Introduce internal "gates" and shortcut paths (like the cell state in LSTMs) to prevent gradients from vanishing or exploding, allowing them to remember information over long sequences.
* **Transformers:** Entirely replace recurrence with self-attention mechanisms, enabling massive parallelization and flawless long-range context tracking.

```python
class lstmbyhand(nn.Module):
    def __init__(self):
        # initalize weights and biases
        super().__init__()

        self.wlr1=nn.Parameter(torch.randn(1,requires_grad=True))
        self.wlr2=nn.Parameter(torch.randn(1,requires_grad=True))
        self.blr1=nn.Parameter(torch.tensor(0.0),requires_grad=True)

        self.wpr1=nn.Parameter(torch.randn(1,requires_grad=True))
        self.wpr2=nn.Parameter(torch.randn(1,requires_grad=True))
        self.bpr1=nn.Parameter(torch.tensor(0.0),requires_grad=True)

        self.wp1=nn.Parameter(torch.randn(1,requires_grad=True))
        self.wp2=nn.Parameter(torch.randn(1,requires_grad=True))
        self.bp1=nn.Parameter(torch.tensor(0.0),requires_grad=True)

        self.wo1=nn.Parameter(torch.randn(1,requires_grad=True))
        self.wo2=nn.Parameter(torch.randn(1,requires_grad=True))
        self.bo1=nn.Parameter(torch.tensor(0.0),requires_grad=True)
    def lstm_unit(self,in_val,long_val,short_val):
        # % of long term memmory to be remebered
        long_rem_percent=torch.sigmoid((self.wlr1*short_val)+(self.wlr2*in_val)+self.blr1)
        #remove comments to check
        #return long_rem_percent
        potential_rem_percent_long = torch.sigmoid((self.wpr2*in_val)+(self.wpr1*short_val)+self.bpr1) #for updating long term memory using short term memory
        #return potential_rem_percent_long
        potential_memory_long = torch.tanh((short_val*self.wp1)+(in_val*self.wp2)+self.bp1)
        updated_long_memory = (long_val*long_rem_percent)+(potential_rem_percent_long*potential_memory_long)
        potential_rem_percent_short = torch.sigmoid((self.wo1*short_val)+(self.wo2*in_val)+self.bo1)
        potential_memory_short = torch.tanh(updated_long_memory)
        updated_short_memory = potential_memory_short*potential_rem_percent_short
        return [updated_long_memory,updated_short_memory]
    def forward(self,price_list):
        long_val = torch.tensor(0.0)
        short_val = torch.tensor(0.0)
        for price in price_list:
            long_val,short_val = self.lstm_unit(price,long_val,short_val)
        return short_val
```

```python
def mse_loss(predict,target):
    squared_errors = (predict-target)**2
    loss = torch.mean(squared_errors)
    return loss
```

```python
model = lstmbyhand()
x_train = torch.tensor([0.1, 0.2, 0.3, 0.4])
y_target = torch.tensor(0.5)
pred = model.forward(x_train)
pred
```

```python
for _ in range(500):
    pred = model.forward(x_train)
    loss = mse_loss(pred,y_target)
    for p in model.parameters():
        p.grad=None
    loss.backward()
    lr = 0.01
    for p in model.parameters():
        p.data -= p.grad*lr
print(loss)
```
