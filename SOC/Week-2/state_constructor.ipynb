```python
import torch

# 1. Load the dictionary from the uploaded file
loaded_state = torch.load('scratch_mlp_mnist_full.pt')

# 2. Reconstruct the variables exactly as they were, mapped directly onto the GPU
W1 = loaded_state['W1'].to('cuda')
b1 = loaded_state['b1'].to('cuda')
W2 = loaded_state['W2'].to('cuda')
b2 = loaded_state['b2'].to('cuda')

# 3. Rebuild your parameters list and restore their gradient tracking settings
parameters = [W1, b1, W2, b2]
for p, req_grad in zip(parameters, loaded_state['parameters_requires_grad']):
    p.requires_grad = req_grad

print("--- MODEL FULLY RESTORED ---")
print("Your trained from-scratch model is loaded onto the GPU and ready for inference!")
```
